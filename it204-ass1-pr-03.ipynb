```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
time = np.arange(0, 2, 0.01)
amp = np.sin(2*np.pi*time)
plt.figure()
plt.plot(time, amp)
plt.title('continuous time signal of y = sin(t)')
plt.xlabel('t');
plt.ylabel('sin(t)')
plt.show()
```

```python
n = np.arange(50);

dt = 0.07/50
x = np.sin(2*np.pi*dt)
plt.xlabel('n');
plt.ylabel('x[n]');
plt.title(r'Plot of DT signal $x[n] = 325 \sin(2\pi 50 n \Delta t)$');
plt.stem(n, x);
```
