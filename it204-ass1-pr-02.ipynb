```python
# Importing necessary packages
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
# function to plot the graph
plt.plot(np.arange(-5, 5), signal.unit_impulse(10, 'mid'))
# labeling the x-axis and y-axis
plt.xlabel('Time')
plt.ylabel('Amplitude')
# to print the graph
plt.show()
```

```python
# declaring the variable n
n = range(-2, 6, 1)
y = []
for i in range(len(n)):
    temp = (1 if n[i]>=0 else 0)
    y.append(temp)


#plotting the graph
plt.stem(n, y)
plt.axis([0, 5.1, -0.1, 1.2])
# to print the graph
plt.show()
```

```python
# declaring the variable n
n = range(-2, 6, 1)
y = []
for i in range(len(n)):
    temp = (n[i] if n[i]>=0 else 0)
    y.append(temp)
    

#Plotting the graph
plt.stem(n, y)
plt.axis([0, 5.1, -0.1, 5.2])
# labeling the x-axis and y-axis
plt.xlabel('  n--->  ')
plt.ylabel('Amplitude ----> ')
# Title of the graph
plt.title('Unit ramp Signal')
# printing the graph
plt.show()
```
