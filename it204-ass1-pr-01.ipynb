```python
# importing required libraries
import matplotlib.pyplot as plt
import numpy as np

# declaring time as a variable and assigning range of time
time = np.arange(0, 10, 0.01)
# variable amp is storing the value of sin(time)
amp = np.sin(time)
plt.figure()
# function to plot the graph
plt.plot(time, amp)
# Title of the graph
plt.title('sin wave')
# labeling the x-axis and y-axis
plt.xlabel('time');
plt.ylabel('amplitude')
# to print the graph
plt.show()
```

```python
# declaring time as a variable and assigning range of time
time = np.arange(0, 2, 0.01)
# variable amp is storing the value of cos(time)
amp = np.cos(2*np.pi*time)
plt.figure()
# function to plot the graph
plt.plot(time, amp)
# Title of the graph
plt.title('cosine wave')
# labeling the x-axis and y-axis
plt.xlabel('time');
plt.ylabel('amplitude')
# to print the graph
plt.show()
```

```python
# importing library
from scipy import signal
# declaring the variable t 
t = np.linspace(0, 1, 500)
# function to plot the graph
plt.plot(t, signal.sawtooth(2 * np.pi * 5 * t))
# Title of the graph
plt.title('sawtooth wave')
# labeling the x-axis and y-axis
plt.xlabel('time');
plt.ylabel('amplitude')
# to print the graph
plt.show
```

```python
# declaring the variable t 
t = np.linspace(0, 1, 500, endpoint=False)
# function to plot the graph
plt.plot(t, signal.square(2 * np.pi * 5 * t))
plt.ylim(-2, 2)
# Title of the graph
plt.title('square wave')
# labeling the x-axis and y-axis
plt.xlabel('time');
plt.ylabel('amplitude')
# to print the graph
plt.show
```

```python
# declaring the variable t 
t = np.linspace(0, 1, 500)
triangle = signal.sawtooth(2 * np.pi * 5 * t, 0.5)
# function to plot the graph
plt.plot(t, triangle)
# Title of the graph
plt.title('Triangular wave')
# labeling the x-axis and y-axis
plt.xlabel('time')
plt.ylabel('amplitude')
# to print the graph
plt.show
```
